```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\user\OneDrive\Desktop\github projects\ecommerce-customer-churn\data\raw\online_retail_09_10.csv",
    encoding="ISO-8859-1"
)

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.head()
```

```python
# Clean and standardize column names
df.columns = (
    df.columns
      .str.replace('ï»¿', '', regex=False)  # remove BOM
      .str.strip()
      .str.lower()
)

# Rename columns to consistent names
df = df.rename(columns={
    'invoiceno': 'invoice',
    'invoicedate': 'invoice_date',
    'unitprice': 'price',
    'customerid': 'customer_id'
})

df.columns
```

```python
df = df.dropna(subset=['customer_id'])
df = df[df['quantity'] > 0]
df = df[df['price'] > 0]

df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df['total_amount'] = df['quantity'] * df['price']
```

```python
import os

os.makedirs("data/processed", exist_ok=True)

df.to_csv("data/processed/clean_transactions.csv", index=False)
```
