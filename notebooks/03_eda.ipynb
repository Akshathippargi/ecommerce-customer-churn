```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/clean_transactions.csv")
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
```

```python
orders_per_customer = df.groupby('customer_id')['invoice'].nunique()

sns.histplot(orders_per_customer, bins=50)
plt.title("Orders per Customer")
plt.show()
```

```python
revenue_per_customer = df.groupby('customer_id')['total_amount'].sum()

sns.histplot(revenue_per_customer, bins=50)
plt.title("Revenue per Customer")
plt.show()
```

```python
snapshot_date = df['invoice_date'].max()
recency = (
    snapshot_date - df.groupby('customer_id')['invoice_date'].max()
).dt.days

sns.histplot(recency, bins=50)
plt.title("Customer Recency (days)")
plt.show()
```
